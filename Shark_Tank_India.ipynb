# **Total amount invested (in Lakhs) by Sharks in different startups.**

```python
AshneerGroverDeal = df[df.AshneerDeal==1]
AnupamMittalDeal = df[df.AnupamDeal==1]
AmanGuptaDeal = df[df.AmanDeal==1]  
VineetaSinghDeal = df[df.VineetaDeal==1]
PeyushBansalDeal = df[df.PeyushDeal==1]
NamitaThaparDeal = df[df.NamitaDeal==1]
GhazalAlaghDeal = df[df.GhazalDeal==1]
```

```python
AshneerInvestment=AshneerGroverDeal['AmountPerShark'].sum()
AnupamInvestment=AnupamMittalDeal['AmountPerShark'].sum()
AmanInvestment=AmanGuptaDeal['AmountPerShark'].sum()
VineetaInvestment=VineetaSinghDeal['AmountPerShark'].sum()
PeyushInvestment=PeyushBansalDeal['AmountPerShark'].sum()
NamitaInvestment=NamitaThaparDeal['AmountPerShark'].sum()
GhazalInvestment=GhazalAlaghDeal['AmountPerShark'].sum()

print("Total amount invested (in Lakhs) by Ashneer Gover :",AshneerInvestment)
print("Total amount invested (in Lakhs) by Anupam Mittal :",AnupamInvestment)
print("Total amount invested (in Lakhs) by Aman Gupta :",AmanInvestment)
print("Total amount invested (in Lakhs) by Vineeta Singh :",VineetaInvestment)
print("Total amount invested (in Lakhs) by Peyush Bansal :",PeyushInvestment)
print("Total amount invested (in Lakhs) by Namita Thapar :",NamitaInvestment)
print("Total amount invested (in Lakhs) by Ghazal Alagh ;",GhazalInvestment)

count=[AshneerInvestment,AnupamInvestment,AmanInvestment,VineetaInvestment,PeyushInvestment,NamitaInvestment,GhazalInvestment]
total=sum(count)

plt.figure(figsize=(8,8))
list=["AshneerInvestment", "AnupamInvestment","AmanInvestment", "VineetaInvestment","PeyushInvestment","NamitaInvestment","GhazalInvestment"]
plt.pie(count, labels=list,autopct=lambda p: 'Rs {:.0f}'.format(p * total / 100))
plt.title('Total investment made by individual shark(in Lakhs)')
plt.show()
```

**Higest investment is done by Aman Gupta i.e. Rs.880 Lakhs and lowest by Ghazal Alagh i.e. Rs. 130 Lakhs.**

```python
AshneerEquity=AshneerGroverDeal['EquityPerShark'].sum()
AnupamEquity=AnupamMittalDeal['EquityPerShark'].sum()
AmanEquity=AmanGuptaDeal['EquityPerShark'].sum()
VineetaEquity=VineetaSinghDeal['EquityPerShark'].sum()
PeyushEquity=PeyushBansalDeal['EquityPerShark'].sum()
NamitaEquity=NamitaThaparDeal['EquityPerShark'].sum()
GhazalEquity=GhazalAlaghDeal['EquityPerShark'].sum()

print("Total percentage of equity owned by Ashneer Gover :", AshneerEquity)
print("Total percentage of equity owned by by Anupam Mittal :", AnupamEquity)
print("Total percentage of equity owned by Aman Gupta :", AmanEquity)
print("Total percentage of equity owned by Vineeta Singh :",VineetaEquity)
print("Total percentage of equity owned by Peyush Bansal :", PeyushEquity)
print("Total percentage of equity owned by Namita Thapar :", NamitaEquity)
print("Total percentage of equity owned by Ghazal Alagh :", GhazalEquity)

count=[AshneerEquity,AnupamEquity,AmanEquity,VineetaEquity,PeyushEquity,NamitaEquity,GhazalEquity]
total=sum(count)

plt.figure(figsize=(8,8))
list=["AshneerEquity", "AnupamEquity","AmanEquity", "VineetaEquity","PeyushEquity","NamitaEquity","GhazalEquity"]
plt.pie(count, labels=list,autopct=lambda p: ' {:.0f} %'.format(p * total / 100))
plt.title('Total percentage of equity owned by Sharks in different startups.')
plt.show()
```
